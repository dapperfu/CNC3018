```python
%aimport GCode
%aimport GRBL
```

```python
cnc = GRBL.GRBL(port="/dev/cnc_3018")
```

```python
def init(feed = 200, laser = 25):
    program = GCode.GCode()
    program.G21() # Metric Units
    program.G91() # Absolute positioning.
    program.G1(F=feed) #
    program.M3(S=laser) # Laser settings.
    return program
```

```python
def end():
    program = GCode.GCode()
    program.M5() # Laser settings.
    return program
```

```python
def square(size=20):    
    program = GCode.GCode()
    program.G1(X=size)
    program.G1(Y=size)
    program.G1(X=-size)
    program.G1(X=-size)
    return program
```

```python
init(laser=255) + square() + end()
```

```python
cnc.run(init(laser=50) + square() + end())
```
